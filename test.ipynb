### Getting  US County density json data

```python
x = df_1.groupby(grp_col).first()
county_df= x.reset_index()
county_df.head()
```

#### I want to change COUNTY FIPS to 5 digit string instead of numerical value.


```python
def po(x):
    x_str = str(x) 
    row_len = len(x_str)
    if row_len<5 and row_len>0:
        extra_prepends = 5 - row_len
        return "0"*extra_prepends + x_str
    return x_str
# df['county_fips'] = df['new_county_fips'].apply(po)
county_df['county_fips'] = county_df['county_fips'].apply(po)
county_df['county_fips']
```
